### 2. (4 points) 
Explain why the optimal market/limit order decision depends in general on quantity available on both sides of the order book in the Parlour model.

On one hand, the quantity avaible on one side of the orderbook increase, the probability of submitting a market order on this side will increase, since it will be easier to fulfill than a limit order as limit orders will be waiting for a long queue.

```python
# e)
def optimal_order_size(h, f, r, xi, sigma, lbda):

    z = (2*h+f+r)/(lbda+h+r)

    return lognorm.ppf(q=z, s=sigma, scale=np.exp(-0.5 * sigma**2 + np.log(xi)))


print("Optimal Order Size on BZX: %.2f" % optimal_order_size(
    h=0.005, f=0.003, r=0.0032, xi=105000, sigma=1, lbda=0.0375))

print("Optimal Order Size on BYX: %.2f" % optimal_order_size(
    h=0.005, f=-0.0003, r=-0.0003, xi=13000, sigma=1, lbda=0.0375))
```

## The Bouchaud-M&eacute;zard-Potters order book approximation

### 4. (15 points)

When $\lambda(u)=\lambda$,  a constant, the BMP order book density becomes

\begin{eqnarray*}
\rho_{BMP}(\Delta) &=& \frac{\lambda}{\nu}\,\left\{e^{-\Delta/{\tilde \Delta}}\,\int_0^{\Delta/{\tilde\Delta}}\,\sinh(u)\,du + \sinh(\Delta/{\tilde \Delta})\,\int_{\Delta/{\tilde\Delta}}^\infty\,e^{-u}\,du\right\} \\
&=&\frac{\lambda}{\nu}\,\left\{1 - e^{-\Delta/{\tilde \Delta}}  \right\},
\end{eqnarray*}

where the characteristic distance $\tilde\Delta$ is given by

$$
\tilde \Delta=\sqrt{\frac{\sigma^2}{2\,\nu}}.
$$

(a) Use the provided `zero_intelligence` module with $\lambda = 1$, $\mu = 10$ and $\nu = \frac15$ to simulate 100,000 events. Count the number of limit orders generated from the simulation and plot the resulting average book shape in red (in a band $\pm 30$ around the mid price).

(b) In Lecture 1, by using dimensional analysis we estimated that the variance $\sigma^2_{DA}$ per unit time should be of the order of

$$
\sigma^2_{DA} = \frac{\mu^2 \,\nu}{\lambda^2}.
$$

Compute the characteristic distance $\tilde \Delta_{DA}$ with this estimate $\sigma^2_{DA}$.

(c) Using the estimate of $\tilde \Delta$ from part (b), code a function to compute the corresponding order book density $\rho_{DA}(\Delta)$.

(d) Superimpose the plot of the density $\rho_{DA}(\Delta)$ (in blue) on your plot of the average book shape.

(e) Dimensional analysis can only be expected to give us an estimate of $\sigma^2$ up to some prefactor of order one.  Amend your function `rho_DA` to use a prefactor in the estimate of $\sigma^2$.

(f) Assuming a pre-factor of $\tfrac12$, superimpose the amended asymptotic density on your plot from part (d).

(g) Guess an accurate formula for $\sigma^2$ in terms of $\mu$, $\lambda$, and $\nu$.


```python
lmda, mu, nu = 1, 10, 0.2
zi = ZI(lmda=lmda, mu=mu, nu=nu)
```

```python
# a) Simulate 100,000 events and plot the resulting order book shape with a band of 30
n_events = 100000
band = 30
event_log, avg_book_shape = zi.generate_events(
    n_events, band=band, logging=True)
```

```python
# Count the number of limit orders generated from the simulation

count = 0
for event, _ in event_log:
    if event == "LB" or event == "LS":
        count += 1
print("Number of limit orders generated: %d" % count)
```

```python
x = [i for i in range(-band, band+1)]
plt.figure(figsize=(12, 8))
plt.plot(x, avg_book_shape, "ro:")
plt.xlabel("Relative price")
plt.ylabel("Quantity")
plt.show()
```

```python
# b) Compute the characteristic distance with the estimator


sig_sqr = mu**2 * nu / (lmda**2)
char_dist = np.sqrt(sig_sqr / (2 * nu))


print("Variance per time unit: %.2f" % sig_sqr)
print("Characteristic distance: %.2f" % char_dist)
```

```python
# c) use the estimace of the characteristic distance to compute the orderbook density rhoDA(\Delta)

def rho_DA(delta, lmda, mu, nu):
    sig_sqr = mu**2 * nu / (lmda**2)
    char_dist = np.sqrt(sig_sqr / (2 * nu))
    return lmda/nu*(1-np.exp(-np.abs(delta)/char_dist))
```

```python
# d) compute the orderbook density rhoZI(\Delta) and plot it together with rhoDA(\Delta)
x = [i for i in range(-band, band + 1)]
plt.figure(figsize=(12, 8))
plt.xlabel("Relative price")
plt.ylabel("Quantity")
plt.plot(x, avg_book_shape, "ro:", label="rhoZI")
rho = [rho_DA(i, lmda, mu, nu) for i in x]
plt.plot(x, rho, "b", label="rhoDA")
plt.legend()
plt.show()
```

```python
# e) compute the prefactor of the orderbook density rhoZI(\Delta) and plot it together with rhoDA(\Delta)
def prefac_rho_DA(pf, delta, lmda, mu, nu):
    sig_sqr = mu**2 * nu / (lmda**2)
    char_dist = np.sqrt(sig_sqr / (2 * nu))
    return pf*lmda/nu*(1-np.exp(-np.abs(delta)/char_dist))
```

```python
# f)
pf = 0.5

x = [i for i in range(-band, band + 1)]
plt.figure(figsize=(12, 8))
plt.xlabel("Relative price")
plt.ylabel("Quantity")
plt.plot(x, avg_book_shape, "ro:", label="rhoZI")
rho = [prefac_rho_DA(pf, i, lmda, mu, nu) for i in x]
plt.plot(x, rho, "b", label="rhoDA with prefactor = %.2f" % pf)
plt.legend()
plt.show()
```

```python
# g) guess the prefactor of the orderbook density rhoZI(\Delta) and plot it together with rhoDA(\Delta)

factors = [1 + 0.01 * i for i in range(0, 21)]
colors = plt.cm.BuPu(np.linspace(0, 1, len(factors)))

res_list = []
x = [i for i in range(-band, band + 1)]
plt.figure(figsize=(12, 8))
plt.xlabel("Relative price")
plt.ylabel("Quantity")
plt.plot(x, avg_book_shape, "ro:", label="ZI")
for i in range(len(factors)):
    rho = [prefac_rho_DA(factors[i], j, lmda, mu, nu) for j in x]
    plt.plot(x, rho, c=colors[i], label=f"DA(f=%.2f)" % factors[i])
    avg_diff_with_ZI = np.mean(
        np.abs(np.array(avg_book_shape) - np.array(rho)))
    error_sq = np.sum((np.array(avg_book_shape) - np.array(rho))**2)
    res_list.append((factors[i], avg_diff_with_ZI, error_sq))

plt.legend()
plt.show()

print("Factor\tAvg. Diff.\tError Squared")
for i in range(len(res_list)):
    print("%.2f\t%.5f\t\t%.5f" %
          (res_list[i][0], res_list[i][1], res_list[i][2]))
```

We iterate over prefactor values from 1 to 1.2 with a step of 0.01 and plot the results. We can compare the results with the plot of the average book shape in red (in a band $\pm 30$ around the mid price) and the plot of the density $\rho_{DA}(\Delta)$ (in blue). In terms of avg diff, the best prefactor is 1.06, in terms of sum of error squares, the best prefactor is 1.09.

So an educated guess for the formula for $\sigma^2$ in terms of $\mu$, $\lambda$, and $\nu$ is 

- $\sigma^2 = \frac{\mu^2 \,\nu}{\lambda^2} * 1.09$ if our target is to minimize the sum of error squares

 and 
- $\sigma^2 = \frac{\mu^2 \,\nu}{\lambda^2} * 1.06$ if our target is to minimize the avg diff.
